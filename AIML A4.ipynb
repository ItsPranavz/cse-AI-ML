```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/house_prices.csv')
df
```

```python
# Check the shape of the data
df.shape
```

```python
# Separatimng the features
x = df["AREA"].values
y = df["PRICE"].values
```

```python
print(x)
```

```python
print(y)
```

```python
# Making x two dimensional
x = x.reshape(50,1)
```

```python
# Splitting test and train data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
# Training the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Making predictions for test data
y_pred = regressor.predict(x_test)
y_pred
```

```python
# Visualizing the results
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'cyan')
```

```python
# Evaluate Model Performance
from sklearn.metrics import r2_score

model_score = r2_score(y_train, regressor.predict(x_train))
print(model_score.round(2))

test_score = r2_score(y_test, regressor.predict(x_test))
print(test_score.round(2))
```
